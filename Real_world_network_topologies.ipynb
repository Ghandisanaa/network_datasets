```python
import networkx as nx
import pandas as pd
```

# Abilene

```python
#US backbone network
G = nx.Graph()  # or DiGraph, MultiGraph, MultiDiGraph, etc
abilene_edges=[(1,2),(2,6),(2,12),(2,5),(12,9),(9,3),(3,6),(6,7),(7,5),(7,4),(5,8),(8,10),(10,4),(10,11),(11,4)]
G.add_edges_from(abilene_edges) 
nx.draw(G)
```

```python
mapping = {1: "Atlanta1", 2: "Atlanta2", 3: "Chicago", 4: "Denver", 5: "Houston", 6: "Indianapolis", 7: "Kansas", 8: "Los Angeles", 9: "New York", 10: "Sunnyvale", 11: "Seattle", 12: "Washington"}
G = nx.relabel_nodes(G, mapping)
```

```python
abilene_adj=nx.adjacency_matrix(G)
abilene_adj=abilene_adj.todense()
df=pd.DataFrame(abilene_adj)
df.to_csv('abilene_adj',index=False)
```
